# House Price Prediction:

This is a simple real-data learning regression method using Python's Scikit-Learn library. '
We will be using the Housing dataset from 'Kaggle', which contains information about different thinks like [Sqrft,Fllors,YearBuilt,Renovated].

We will follow these steps:
1. Load the dataset.
2. Explore the dataset.
3. Preprocess the data (cleaning, etc.).
4. Train a simple linear regression model.
    WITH ONLY ONE VARIABLE FOR NOW
5. Predict the Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Let's Load the dataset
df = pd.read_csv('./Data/HousePriceData.csv')
df.head()
```

```python
# Information about The Whole Data
df.info()
```

```python
# Let's Choose Only one Independent Variable For Noe
# so, let's Choose
x = df[['sqft_living']]
y = df['price']
```

```python
# Let's Plot the Data
plt.scatter(x, y, color='blue')
plt.title('Square Footage vs Price')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.show()
# Splitting the dataset into the Training set and Test set
```

```python
# Lets Bring The Model First
model = LinearRegression()
```

Train the Model, So That It Can Make a Best Fit Line
On x -> SQRFT LIVING, y -> PRICE


```python
model.fit(x, y) # The model is trained 
```

```python
# Lets Preduct On Sqrt LIVING = 4500
predicted_value = model.predict([[4500]])
predicted_value
```
