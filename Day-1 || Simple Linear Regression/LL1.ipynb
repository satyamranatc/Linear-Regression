```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
StudyHrs = np.array([1,2,3,4,5,6]) # Independent Variable
Marks = np.array([85,92,88,95,90,93]) # Dependent Variable
```

```python
df = pd.DataFrame({
    'Study Hours': StudyHrs,
    'Marks': Marks
})
df
```

```python
# For Better visualization Lets Plot
plt.plot(df['Study Hours'], df['Marks'],color='blue', marker='x', linestyle='--')
plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

```python
# Now See This Picture To Understand The Best Fit Line On This Data
```

![Pic](./Data/BestFitLine.jpg)

```python
# Lets Create a ML Model Which Create This 
# Best Fit Line So That,
# We Can Predict Something
```

```python
model = LinearRegression() # Assign The ML Model(Algo/Math) in "model" Variable
```

```python
# Lets Train the Model
model.fit(df[['Study Hours']], df['Marks']) # This Will Learn The Coefficients(M/Slope)
```

```python
# Now The Model Is Trained On Previous Data
# Its Can Predict Something New
```

```python
# Lets Ask The Model 
# If a Person Is Studing 3.5 Hrs
#  What is Expected Marks
model.predict([[3.5]]) #3.5 Hrs
```
